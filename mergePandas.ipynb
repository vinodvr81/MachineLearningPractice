```python
data5 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
   

data6 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 

fdf = pd.DataFrame(data5)
 

fdf1 = pd.DataFrame(data6) 
  
 
print(fdf, "\n\n", fdf1) 
```

```python
res13 = pd.merge(fdf, fdf1, on='key')
 
res13
```

```python
data7 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
   

data8 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 


dtf = pd.DataFrame(data7) 
 
dtf1 = pd.DataFrame(data8) 
  
print(dtf, "\n\n", dtf1)
```

```python
res14 = pd.merge(dtf, dtf1, on=['key', 'key1'])
 
res14
```

```python
res15 = pd.merge(dtf, dtf1, how='left', on=['key', 'key1'])
 
res15
```

```python
res16 = pd.merge(dtf, dtf1, how='outer', on=['key', 'key1'])
 
res16
```

```python
res17 = pd.merge(dtf, dtf1, how='inner', on=['key', 'key1'])
 
res17
```

```python
rng = pd.DataFrame() 
rng['date'] = pd.date_range('1/1/2011', periods = 72, freq ='H') 
  
# Print the dates in dd-mm-yy format 
rng[:5] 
  
# Create features for year, month, day, hour, and minute 
rng['year'] = rng['date'].dt.year 
rng['month'] = rng['date'].dt.month 
rng['day'] = rng['date'].dt.day 
rng['hour'] = rng['date'].dt.hour 
rng['minute'] = rng['date'].dt.minute 
  
# Print the dates divided into features 
rng.head(3)
```
